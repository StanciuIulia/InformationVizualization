To create the visualization two .csv databases were used. 
One was created by hand and it includes data on the 10 departments: acronym, adress, website, number of actual members. 
The second one was obtained by refining and augmenting the database provided in the lab instructions. 

***The steps of the data refinement and augmentation were the following:***

1.   Using a facet, the people who were not part of a departament were removed from the database. Institutes or centers clasified as departments were also removed, leaving us with 30 departments after this first step. 
2.   The date of publication was reduced to year by am expression change to simplify the repesentations. 
3.    To ensure the consistency of the data, the format of the ISSN was checked. Some discrepencies from the normal format were found and were treated in various ways: if only the dash character was missing, it was added in its place, if the ISSN was includent, but we were able to find the publication and its ISSN online, the number was changed to the correct one. In some special cases, the whole entry was deleted. 
4.   The consistency between authors and author ID was checked.
5.   The 10 departments needed for this project were chosen considering geographical position, subject diversity, size and number of articles published.
6.   The departments' names were translated to english as they can be found on the UPC website. 





***Analysis of the project:***

1.   The .html page in made out of a header with the title, five buttons and two view points for the visualizations. 

2.   The buttons have the following functionalities:

*   "See Map" - helps the user to see the interactive map.
*   "Visit the University Website" - opens the university website in another tab.
*   "Project Information" - gives important information about the project. The information is displayed in the main page, where the map is in the first in the begining. The user can use the scroll bar to scroll through information. 
*   "See department size vs number of publications" - to see the bubble chart and table containing information about total number of members and articles written by each department. 
*   "Compare department productivity" - to see the coordonated views charts and compare two ore more by clicking on the colored rectangular buttons on the top of the left part of the screen.

3.   The project makes use of an interactive map with two types of markers with different funcionalities implemented in the click function and hover function. The map center is the average of the coordinates of the 10 departments' locations. The small map in the bottom left of the big map helps orientation and calibration. 

4.   By clicking on the red pin points, some basic information about the department is provided. The website of each department opens in a different tab when the link is pressed.

5.   On the other hand, when clicking on the blue circles a box plot with points representing authors of that department is provided. The plot depicts the number of articles written by each author of the department in each year. This part of the visualization helps us see which was the most productive year for some authors, as well as an average number of articles per author by year. For example, the Department of Mathematics has a steady number of articles per author, but in years 2010, 2012 and 2019 some authors with over 12 articles written in that year can be noticed. Some other aspects can be observed for the Department of Mining, Industrial and ITC Engineering in Manresa where the year 2018 has been found the most productive in terms of articles per member, with an average of 3, almost double compared with other years.

6.    For the second part of the vizualization, two color coded graphs can be interchanged. The first one contains a bar chart illustrating the each department's productivity by year in terms of written articles. The bottom one is a line chart, showing the evolution in the number of written articles by the department with respect to time (from 2010 to a specific year). Comparisons between departments can be made by selecting the wanted ones from the colored chart placed on top of the stacked bar chart. These coordinated views charts placed side by side provide an understanding of the yearly production as well as the productivity trend for each deprtment. One interesting aspect which can be seen using this part of the visualization is comparing one of the most productive departments, such as Department of Mathematics with a less productive part of UPC in terms of publications, for example Department of Nautical Science and Engineering. It can be easily seen that MAT has written an average of 200 articles per year, while CEN aproximately 15-20. The line chart shows an even larger discrepancy in productivity.
